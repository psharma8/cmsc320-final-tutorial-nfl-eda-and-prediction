Data Collection <br>
----------------------- <br>
There are csv files where each csv file contains nfl play by play details of each game <br> 
for regular season of a single year <br>
We collect data from ten individual csv files to get information about nfl play by play detail <br>
from 2009 to 2017 and part of 2018 that are played

```python
# Libraries and classes that are going to be used through out the project
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy.polynomial.polynomial as poly
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
file_str = 'regular_season/reg_pbp_'
start_year = 2009
csv_str='.csv'
list_for_df=[]
colnames = ['game_id', 'home_team', 'away_team', 'posteam','defteam', 'game_date', 'qb_kneel', 
            'field_goal_result','td_team','total_home_score', 'total_away_score', 
            'incomplete_pass', 'interception', 'qb_hit', 'sack', 'touchdown','pass_touchdown',
            'rush_touchdown', 'return_touchdown', 'fumble', 'complete_pass',   'penalty_team',
            'penalty_yards']
df=pd.DataFrame(columns=colnames)
for i in range(10):
    year=start_year+i
    year_str = str(year)
    file_name = file_str+year_str+csv_str;
    # data in csv file has more than 200 columns so to limit the reading data pass usecols which
    # save time and space
    df1 = pd.read_csv(file_name,usecols=colnames) 
#     df1=df1[colnames]
    df1_len = len(df1)
    year_list = np.full((df1_len, 1), year)
    df1['year']=year_list
    df=df.append(df1)
    
# change datatype of year to integer
df['year']=df['year'].astype(int)

# fix the name of teams on data 
# Since Rams and Chargers moved to LA from thier previous place , their names on data have been
# changed Also Jacksonville is represented by 'JAC' and later it's changed to 'JAx'. 
# we replace all 'STL' to 'LA', 'SD' to 'LAC', and 'JAC' to 'JAX'
df = df.replace('STL', 'LA')
df = df.replace('SD','LAC')
df = df.replace('JAC','JAX')


df.head()
```

Data Wrangling Part(1) : <br>
--------------------------------- <br>
Data we collect contains play by play details (Dirty Data) which cannot be used until <br> 
we process the data and make it tidy and usable <br>
After wrangling process we will have only two rows for each game with all necessary detail of each team who <br>
played game against each other. It will reduce the number of multiple rows into two rows and make it more easier to <br>
use, read and visualize the data.<br>
During this wrangling process we create two new dataframes <br>
df      :  for ploting and analyzing data <br>
pred_df : dataframe is for prediction of future games using logistic regression <br>


```python
# col_names contains columns for which we are going to get values for each team from 
# all games they played
col_names=['game_id','year', 'team', 'away' , 'total_scores', 'lose_win','penalty_yards',
           'qb_kneel', 'fg_missed' ,'fg_made' ,'incomplete_pass' , 'complete_pass' , 'qb_hit' ,
           'fumble' , 'sack', 'interception','year']
pred_col_names = ['year','home_team','away_team','home_team_win']
pred_list = []
# first group the rows by year 
for year_key,main_grp in df.groupby(['year']):
    # once we split the data in group by year we need to split them in sub-group for each game
    # because there multiple rows for one single game as the data is for play by play
    for key, grp in main_grp.groupby(['game_id']): 
        # get the total point of each team and their game type( home or away)
        total_away = grp['total_away_score'].max()
        total_home = grp['total_home_score'].max()
        home_team_list = grp.home_team.unique()
        away_team_list = grp.away_team.unique()
        home_team = grp.home_team.unique()[0]
        away_team = grp.away_team.unique()[0]
        away_list = []
        home_list = []
        game_list = []
        home_list.append(key);away_list.append(key)
        home_list.append(year_key);away_list.append(year_key);away_list.append(away_team);
        home_list.append(home_team);home_list.append(0);away_list.append(1);home_list.append(total_home);
        away_list.append(total_away);game_list.append(year_key); game_list.append(home_team);
        game_list.append(away_team);

        # categorize each game as hame or away for each team and count total_scores and put 
        # the value for each team based on who win or lose or both tie 
        # these information will be used to evauluate overall team performance and for prediction

        if(total_away > total_home):
            home_list.append(-1)
            away_list.append(1)
            game_list.append(0)
        elif(total_home > total_away):
            home_list.append(1)
            away_list.append(-1)
            game_list.append(1)
        else:
            home_list.append(0)
            away_list.append(0)
            game_list.append(0)
        #     penalty stats
        for sub_key, sub_grp in grp.groupby(['penalty_team']):
            penalty_yards =sub_grp['penalty_yards'].sum()
            if(sub_key == home_team):
                home_list.append(penalty_yards)
            elif(sub_key == away_team):
                away_list.append(penalty_yards)

        # count penalty_yards, qb_kneel, field goal missed, field goal made, incomplete passes,
        # complete passes, and quarterback hit 
        # penalty yards will be used along some other information to evaluate overall team performance
        # and prediction, while all other information will be used to evaluate offense and to predict       
        for sub_key, sub_grp in grp.groupby(['posteam']):
            qb_kneel = sub_grp['qb_kneel'].sum()
            fg_missed = len(sub_grp[sub_grp.field_goal_result == 'missed'])        
            fg_missed += len(sub_grp[sub_grp.field_goal_result == 'blocked'])        
            fg_made = len(sub_grp[sub_grp.field_goal_result == 'made'])        
            incomplete_pass =sub_grp['incomplete_pass'].sum()        
            complete_pass =sub_grp['complete_pass'].sum()        
            qb_hit = sub_grp['qb_hit'].sum()        
            if(sub_key == home_team):         
                home_list.append(qb_kneel)        
                home_list.append(fg_missed)
                home_list.append(fg_made)
                home_list.append(incomplete_pass)
                home_list.append(complete_pass)
                home_list.append(qb_hit)
            elif(sub_key == away_team): 
                away_list.append(qb_kneel)
                away_list.append(fg_missed)
                away_list.append(fg_made)
                away_list.append(incomplete_pass)
                away_list.append(complete_pass)
                away_list.append(qb_hit)
                
           # count number of fumbles, sacks, interceptions per game for each team
            # these information will be used to evaulate quality of defense and for prediction 
        for sub_key, sub_grp in grp.groupby(['defteam']):
            fumble = sub_grp['fumble'].sum()
            sack = sub_grp['sack'].sum()
            interception = sub_grp['interception'].sum()
            if(sub_key == home_team): 
                home_list.append(fumble)
                home_list.append(sack)
                home_list.append(interception)
            elif(sub_key == away_team): 
                away_list.append(fumble)
                away_list.append(sack)
                away_list.append(interception)
                
        list_for_df.append(away_list)
        list_for_df.append(home_list)
        pred_list.append(game_list)

        
pred_df = pd.DataFrame(columns=pred_col_names)
pred_df=pd.DataFrame(pred_list,columns=pred_col_names)
# now we have data which contains all information of each team for every game they played in
# regular seasons from 2009 to 2017 and some part of the 2018
col_names=['game_id','year', 'team', 'away' , 'total_scores', 'lose_win','penalty_yards', 'qb_kneel',
           'fg_missed' , 'fg_made' , 'incomplete_pass' , 'complete_pass' , 'qb_hit' , 'fumble' ,
           'sack','interception']
df=pd.DataFrame(columns=col_names)
df=pd.DataFrame(list_for_df,columns=col_names)
df.head()
```

Data Wrangling Part(2) : <br>
--------------------------------- <br>
In this part, for one of our dataframe (df), we group dataframe by year and  summarise rows with detail of all games for <br>
each team into one row so we will have data where each row contains information of each team per year from 2009 to 2018

```python
summary_list = []
col_names = ['year','team','away_win','home_win','tie','lose','total_scores','penalty_yards',
             'qb_kneel','fg_missed','fg_made','incomplete_pass','complete_pass','qb_hit',
             'fumble','sack','interception']
for key, grp in df.groupby(['year']): 
    for sub_key, sub_grp in grp.groupby(['team']):
        lst = []
        
        # count total number of away-win, home-win, tie, lose, scores,penalty yards concede,
        # quarter back kneeel,field goal missed, field goal made, incomplete pass, pass completed,
        # quarterback got hit, ball fumbled, quarterback got sacked,pass intercepted
        away_win = len(sub_grp.loc[(sub_grp['away'] == 1) & (sub_grp['lose_win']==1)])
        home_win = len(sub_grp.loc[(sub_grp['away'] == 0) & (sub_grp['lose_win']==1)])
        tie = len(sub_grp[sub_grp.lose_win == 0])
        lose = len(sub_grp[sub_grp.lose_win == -1])
        total_scores = sub_grp['total_scores'].sum()
        penalty_yards = sub_grp['penalty_yards'].sum()
        qb_kneel = sub_grp['qb_kneel'].sum()
        fg_missed = sub_grp['fg_missed'].sum()
        fg_made = sub_grp['fg_made'].sum()
        incomplete_pass = sub_grp['incomplete_pass'].sum()
        complete_pass = sub_grp['complete_pass'].sum()
        qb_hit = sub_grp['qb_hit'].sum()
        fumble = sub_grp['fumble'].sum()
        sack = sub_grp['sack'].sum()
        interception = sub_grp['interception'].sum()
        lst=[key,sub_key,away_win,home_win,tie,lose,total_scores,penalty_yards,qb_kneel,
             fg_missed,fg_made,incomplete_pass,complete_pass,qb_hit,fumble,sack,interception]
        summary_list.append(lst)
df=pd.DataFrame(summary_list,columns=col_names)
d=df.loc[(df['year'] == 2018) & (df['team']=='BAL')]
df.head()
```

<br><br>
in next step, the data is standardized so all value we have for away-win, home-win, tie, lose are in ratio to total games <br>
while all other values of each columns are the average (the total / total games played).<br><br>

```python
# Get weighted score for each column using the ratio
df['total_win'] = df['away_win']+df['home_win']
df['total_fg'] = df['fg_missed']+df['fg_made']
df['total_pass'] = df['complete_pass']+df['incomplete_pass']
df['total_game'] = df['away_win']+df['home_win']+df['tie']+df['lose']
# denominator cannot be zero so first set total to 1 if it's 0 before getting ratio
df['total_fg'] = np.where((df['total_fg']==0),1,df.total_fg)
df['total_pass'] = np.where((df['total_pass']==0),1,df.total_pass)
df['total_game'] = np.where((df['total_game']==0),1,df.total_game)

# overall team performance
df['win_ratio'] = df['total_win']/df['total_game']
df['away_win_ratio'] = df['away_win']/df['total_game']
df['home_win_ratio'] = df['home_win']/df['total_game']
df['tie_ratio'] = df['tie']/df['total_game']
df['lose_ratio'] = df['lose']/df['total_game']
df['total_scores_ratio'] = df['total_scores']/df['total_game']
df['penalty_ratio'] = df['penalty_yards']/df['total_game']

# offensive performance
df['qb_kneel_ratio'] = df['qb_kneel']/df['total_game']
df['fg_ratio'] = df['fg_made']/df['total_fg']
df['completion_ratio'] = df['complete_pass']/df['total_pass']

# defensive performance => fumble, qb_hit, sack, interception
df['fumble_ratio'] = df['fumble']/df['total_game']
df['qb_hit_ratio'] = df['qb_hit']/df['total_game']
df['sack_ratio'] = df['sack']/df['total_game']
df['interception_ratio'] = df['interception']/df['total_game']

df.drop(['fg_made', 'fg_missed','complete_pass','incomplete_pass','total_pass','total_fg'], axis=1,inplace=True)
df.drop(['penalty_yards','away_win', 'home_win','tie','lose','total_scores','total_win','qb_kneel','total_game'], axis=1,inplace=True)
df.drop(['fumble', 'qb_hit','sack','interception'], axis=1,inplace=True)


df_len = len(df)
chmp_list = np.full((df_len, 1), 0)
df['champion'] = chmp_list
lst_of_champs=[(2009, 'NO'), (2010, 'GB'), (2011, 'NYG'),(2012, 'BAL'), (2013, 'SEA'),
                (2014, 'NE'),(2015, 'DEN'), (2016, 'NE'), (2017, 'PHI')]

for yr_tm in lst_of_champs:
    (year,team) = yr_tm
    df['champion'] = np.where((df['year']==year)&(df['team']==team),1,df.champion)
df.head()
```

<b>Prediction of next coming fixtures and whether Ravens will make it to the either Playoff or wildcard-entry</b>
<br>
Libraries from sklearn is used to get prediction <br>
Inependent Variables are (X) : <br>
home-team <br>
away-team <br>
home game win to total games played ratio (total home game wins / total games played) of home team <br>
average total scores (total scores / total games played) of home game <br>
away game win to total games played ratio (total away game wins / total games played) of away team <br>
average total scores (total scores / total games played) of away team <br>
<br>
Dependent Variable (Y) : <br>
home team win (binary if home win then 1 else 0)  here a chance of draw is ignored, and count it as 0. It's very <br>
unlikely to have draw and therefore winning is more important to get to the playoff or wildcard round. Hence, we are <br>
only interested to figure out whether team win or lose (draw is counted as lose)<br>
<br><br>
Now we are going to add more information to dataframe for prediction (pred_df) which we created <br>
in Data Wrangling Part 1


```python
# col_names=['year','home_team','away_team','home_win_team','avg_ht_hw','avg_at_aw','avg']
col_names=['year','home_team','away_team','home_team_win','avg_home_win_ht','avg_total_scores_ht']
new_pred_df = pd.DataFrame(columns=col_names)
for key, grp in pred_df.groupby(['year','home_team']):
    row = df.loc[(df['year'] == key[0]) & (df['team'] == key[1])]
    x=row.iloc[0]['home_win_ratio']
    y=row.iloc[0]['total_scores_ratio'] #  for home_team
    l = len(grp)
    df1=grp.copy()
    df1['avg_home_win_ht'] = [x]*l
    df1['avg_total_scores_ht'] = [y]*l
    new_pred_df = new_pred_df.append(df1)

new_pred_df.head()
    
col_names.append('avg_away_win_at');col_names.append('avg_total_scores_at')
pred_df = new_pred_df
new_pred_df = pd.DataFrame(columns=col_names)
for key, grp in pred_df.groupby(['year','away_team']):
    row = df.loc[(df['year'] == key[0]) & (df['team'] == key[1])]
    x=row.iloc[0]['away_win_ratio']
    y=row.iloc[0]['total_scores_ratio'] #  for away team
    l = len(grp)
    df1=grp.copy()
    df1['avg_away_win_at'] = [x]*l
    df1['avg_total_scores_at'] = [y]*l
    new_pred_df = new_pred_df.append(df1)

new_pred_df=new_pred_df[['year','home_team','away_team','avg_home_win_ht','avg_total_scores_ht',
                         'avg_away_win_at','avg_total_scores_at','home_team_win']]
new_pred_df.head()
```

$$ Column-Name \space Detail  $$
<br>
avg_home_win_ht : total home game win / total games played for home team <br>
avg_total_scores_ht : average total scores per game for home team <br>
avg_away_win_at : total away win / total games played for away team <br>
avg_total_scores_at : average total scores for away team <br>
home_team_win : if home team win then 1 else 0


In next step, we are training logistic regression model using our data so we can use it to predict future games (week 12)

```python
my_df=new_pred_df.copy()
le = preprocessing.LabelEncoder()
le.fit(my_df['home_team'])

# Since we cannot use string variables we must change 
# columns with string variable to continuous variables
# therefore we LabelEncoder to transform columns with
# names of home-team and names of away-team from strings
# continuous variable (in this case it's strings to integers)
my_df['home_team'] = le.transform(my_df['home_team'])
my_df['away_team'] = le.transform(my_df['away_team'])

# get dummy variable to use data to train model
home_team = pd.get_dummies(my_df['home_team'])
away_team = pd.get_dummies(my_df['away_team'])
my_df = pd.concat([my_df,home_team,away_team],axis=1)

# Since we already concatenated home-team and away-team
# we can drop two original columns
my_df.drop('home_team',axis=1,inplace=True)
my_df.drop('away_team',axis=1,inplace=True)

# X contains all (real) independent variables that are used to
# predict the result
X=my_df.drop('home_team_win',axis=1)

# y is the column which contains (real) values we want to predict
y=my_df['home_team_win']

# the above line of code gives y as list of object datatype so
# we change list of object datatype to list of integer
y = list(map(int, y))

# split data into ~2/3 and ~1/3 to train and test data we have
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)

# initialize model (logistic regression model) and train using train data
model = LogisticRegression()
model.fit(X_train,y_train)

# once we trained the model, get predictions from X_test
predictions = model.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
```

```python
accuracy_score(y_test,predictions)
```

The above score is accuracy score wich means the prediction we get is (score value) * 100% chance that the prediction is right.
